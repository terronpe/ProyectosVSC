```python
import mysql.connector
from mysql.connector import errorcode

try:
  cnx = mysql.connector.connect(user='root',
                                password='terronpe',
                              host='localhost',
                              port = "3306",
                              database='world')
except mysql.connector.Error as err:
  if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
    print("Something is wrong with your user name or password")
  elif err.errno == errorcode.ER_BAD_DB_ERROR:
    print("Database does not exist")
  else:
    print(err)
else:
  cnx.close()
```

```python
import mysql.connector

# Establecer la conexión a la base de datos
conexion = mysql.connector.connect(
    host="localhost",
    user="root",
    password="terronpe",
    database="alextestDB"
)

# Crear un cursor para ejecutar consultas SQL
cursor = conexion.cursor()

# Consulta SQL para seleccionar todos los registros de la tabla "personas"
consulta = "SELECT * FROM personas"

# Ejecutar la consulta SQL
cursor.execute(consulta)

# Obtener todos los registros como una lista de tuplas
registros = cursor.fetchall()

# Imprimir los resultados
for registro in registros:
    print("ID:", registro[0])
    print("Nombre:", registro[1])
    print("Registro Alfanumérico:", registro[2])
    print("Edad:", registro[3])
    print("Año de Nacimiento:", registro[4])
    print("Lugar de Nacimiento:", registro[5])
    print("Comida Favorita:", registro[6])
    print()

# Cerrar el cursor y la conexión
cursor.close()
conexion.close()
```
